# Apple Historical Weekly Data
This notebook retrieves the last 10 years of weekly data for Apple (AAPL) using the [yfinance](https://github.com/ranaroussi/yfinance) library.

```python
import yfinance as yf

# Download last 10 years of weekly data for Apple
apple = yf.download('AAPL', period='10y', interval='1wk')
apple
```
